```python
from sentence_transformers import SentenceTransformer,util
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint
```

```python

model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
import pandas as pd
df = pd.read_csv("./sample.csv")
```

```python
words=[]
for i in df['title_keyword']:
        words.append(str(i))
```

```python
sentence_embeddings = model.encode(words)
```

```python
keyword = input("Enter Keyword:")
key_embedding = model.encode(keyword)
```

```python
similarity=[]
for i in range(len(words)):
    similarity.append(cosine_similarity(key_embedding.reshape(1, -1),
           sentence_embeddings[i].reshape(1, -1)))
```

```python
list1 = []
output=[]
for i in range(len(similarity)):
    list1.append((words[i],similarity[i][0][0]))
    
```
